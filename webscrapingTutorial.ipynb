### 1. Modularize the URL for programmatic access

```python
year = "2024"
stat = "per_game"
url = "https://www.basketball-reference.com/leagues/NBA_" + year + "_" + stat + ".html"
print(url)
```

Example: lets programmatically return a list of URLs given a list of years

```python
years = [2020, 2021, 2022, 2023, 2024]

for i in years:
    url = "https://www.basketball-reference.com/leagues/NBA_" + str(i) + "_" + stat + ".html"
    print(url)
```

### 2. Read HTML webpage in pandas

```python
import pandas as pd
```

Using pd.read_html() function to read HTML webpage

```python
df = pd.read_html(url, header = 0)
df
```

Use len(df) to check how many tables there are in the webbage

```python
len(df)
```

Use df[i] to select a particular table. Select the first table

```python
df2024 = df[0]
print(df2024)
```

### 3. Data Cleaning 

In this NBA example, the table header repeats every 20 players.

When Age column is selected, we are looking for the string "Age", because this column should only have numerical values. When we see "Age" instead of a number, we will remove the whole row. 

```python
df2024[df2024.Age == "Age"]
```

Use the df.drop() function to drop all rows that have "Age"

```python
df2024new = df2024.drop(df2024[df2024.Age == "Age"].index)
```

Compare the dimensions of before and after (rows, columns)

```python
print(df2024.shape)
print(df2024new.shape)
```

Display simple histogram

```python
""" 
pd.to_numeric converts the column "PTS" into numeric values from its original string type.
errors="coerce" specifies how to handle errors. In this case, values that cannot be converted to numeric will be set to NaN
"""
df2024new["PTS"] = pd.to_numeric(df["PTS"], errors="coerce") 
sns.histplot(df2024new,
             x="PTS",
             bins=(list(range(0, 41, 5))),
             kde=False
             )
```
